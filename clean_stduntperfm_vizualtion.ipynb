```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
```

```python
stdnt_perfom=pd.read_csv("clnd_stdnt_prfrm.csv")
stdnt_perfom.head()
```

```python

plt.bar(stdnt_perfom["Hours Studied"],stdnt_perfom["Pervious Score"])
plt.xlabel("Hours Studied")
plt.ylabel("Pervious Score")
plt.title("perfomance score")
plt.show
```

```python
values=[stdnt_perfom["Hours Studied"].mean(),stdnt_perfom["Pervious Score"].mean(),stdnt_perfom["final Scores"].mean()]
labels=["studied hours","previous marks","annual marks"]
explodes=(0.3,0,0)
plt.pie(values,labels=labels,autopct="%1.1f%%",startangle=90,explode=explodes)
plt.title("avarge comparsion pie charts")
plt.axis("equal")
```

```python
sns.scatterplot(data=stdnt_perfom,x="Hours Studied",y="final Scores")
plt.title("study hours vs final marks")
plt.show()
```

```python
sns.histplot(stdnt_perfom["final Scores"],kde=True)
plt.title("Annual mark distribution")
plt.show()
```
